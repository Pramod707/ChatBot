```python
from pinecone import Pinecone,ServerlessSpec
index_name ="hybrid-search-langchain-pinecone"
 ##initializing pinecone
pc = Pinecone(
    api_key=apikey)
##create index
if index_name not in pc.list_indexes().names():
    pc.create_index(
    name = index_name,
    dimension = 384,
    metric = "dotproduct",
    spec = ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    ) 
    )
    
    print(pc.list_indexes().names())
```
